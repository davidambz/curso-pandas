# Ordenação

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/customers.csv", sep=";")

# Ordenando do maior para o menor
# O ascending padrão é true e o implace faz com que o dataframe seja alterado
df.sort_values(by="Points", ascending=False, inplace=True)
df.rename(columns={"Name": "Nome", "Points": "Pontos"}, inplace=True)
df
```

```python
df = pd.read_csv("../data/customers.csv", sep=";")

# Outra forma de fazer o tópico acima, mas encadeado
df = (
    df.sort_values(by="Points", ascending=False)
    .rename(columns={"Name": "Nome", "Points": "Pontos"})
    )
df
```

```python
df = pd.read_csv("../data/customers.csv", sep=";")

# Fazendo o sort por pontos e depois por ordem alfabetica
df.sort_values(by=["Points", "Name"], ascending=[False, True])
```

```python
# Juntando os dois ultímos exemplos
df = (
    df.sort_values(by=["Points", "Name"], ascending=[False, True])
    .rename(columns={"Name": "Nome", "Points": "Pontos"})
    )
df
```
