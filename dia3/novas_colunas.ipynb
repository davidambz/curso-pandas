# Criando novas colunas

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/customers.csv", sep=";")
df
```

```python
# Criando uma nova coluna com o dobro de pontos
df["Points_double"] = df["Points"] * 2
df
```

```python
# Criando uma nova coluna fazendo operações entre colunas
df["Points_ratio"] = df["Points_double"] / df["Points"]
df
```

```python
# Criando uma coluna com o resultado constante
df["Constante"] = None
df
```

```python
# Logaritimo
df["Points_log"] = np.log(df["Points"])
df
```

```python
# Uppercase
df["Name"].str.upper()
```

```python
# Aplicando uma função própria
def get_first(nome:str):
    nome = nome.upper()
    return nome.split("_")[0]

df["Name"].apply(get_first)
```

```python
# Exemplo acima usando lambda
df["Name"].apply(lambda x: x.upper().split("_")[0])
```

```python
def intervalo_pontos(pontos):
    if pontos < 2500:
        return "baixo"
    elif pontos < 3500:
        return "medio"
    else:
        return "alto"
    
df["faixa_pontos"] = df["Points"].apply(intervalo_pontos)
df
```

```python
# Aplicando função no dataframe, e não na série
data = {
    "nome": ["Teo", "Nah", "Maria", "Lara"],
    "recencia": [1,30,10,45],
    "valor":[100,2000, 15, 500],
    "frequencia":[2, 5, 1, 15]
}

df_crm = pd.DataFrame(data)

def rfv(row):
    
    nota = 0
    
    if row['recencia'] <= 10:
        nota += 10
    elif 10 < row['recencia'] <= 30:
        nota += 5
    elif row['recencia'] > 30:
        nota += 0

    if row['valor'] > 1000:
        nota += 10
    elif 100 <= row['valor'] < 1000:
        nota += 5
    elif row['valor'] < 100:
        nota += 0

    if row['frequencia'] > 10:
        nota += 10
    elif 5 <= row['frequencia'] < 10:
        nota += 5
    elif row['frequencia'] < 5:
        nota += 0
    
    return nota

df_crm["RFV"] = df_crm.apply(rfv, axis=1)
df_crm
```
